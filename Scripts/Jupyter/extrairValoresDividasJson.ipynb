```python
dict = {'nomeCliente': [], 'categoria':[], 'tipoDivida': [], 'grupo': [], 'valor da dívida': [], 'nomeDivida': []}
listaCategoria = []
listaTipo = []
listaGrupo = []
listaValor = []
listaDivida = []

#for itemJson in listaJson:
with open(listaJson[0], encoding='utf-8') as meu_json:
    dados = json.load(meu_json)
    dadosJson = dados['scr_data']

    for x in dados['signers']:
        nome = x['name']
        email = x['email']
    
    for i in dadosJson:
        for j in i['operation_items']:
            categoria = j['category_sub']['category']['category_description']
            tipo = j['due_type']['description']
            grupo = j['due_type']['due_type_group']
            valor = j['due_value']
            divida = j['category_sub']['description']

            if grupo == "A vencer":
                listaCategoria.append(categoria)
                listaTipo.append(tipo)
                listaGrupo.append(grupo)
                listaValor.append(valor)
                listaDivida.append(divida)


#dadosJson
```

```python
#Função DE-PARA
def deParaDividas(x):
    if 'Cartão' in x or 'cartao' in x or 'cartão' in x or 'Cartao' in x:
        x = 'Cartão de Crédito'
    elif 'capital' in x or 'Capital' in x:
        x = 'Capital de Giro'
    elif 'cheque' in x or 'Cheque' in x:
        x = 'Cheque Especial'
    elif 'Conta garantida' in x or 'conta garantida' in x:
        x = 'Cheque Especial'
    elif 'sem consignação' in x:
        x = 'Crédito Pessoal s/ Consignação'
    elif 'Crédito pessoal' in x or 'crédito pessoal' in x:
        x = 'Crédito Pessoal'
    elif 'adiantamentos a depositantes' in x or 'desconto de duplicatas' in x:
        x = 'Desconto de duplicatas'
    elif 'aquisição de bens – outros bens' in x:
        x = 'Financiamento Outros Bens'
    elif 'aquisição de bens – veículos automotores' in x:
        x = 'Financiamento de Automóvel'
    elif 'financiamento habitacional – SFH' in x:
        x = 'Financiamento de Imóvel - SFH'
    elif 'financiamento habitacional – exceto SFH' in x:
        x = 'Financiamento de Imóvel'
    elif 'contratado e não utilizado' in x or 'custeio e pré-custeio' in x or 'microcrédito' in x or 'outros empréstimos' in x or 'recebíveis adquiridos' in x:
        x = 'Outros empréstimos'
    elif 'financiamento de projeto' in x or 'outros financiamentos' in x or 'Outros financiamentos' in x:
        x = 'Outros financiamentos'
    elif 'Descontos' in x or 'Outros Limites' in x or 'avais e fianças honrados' in x or 'garantias prestadas' in x or 'desconto de cheques' in x:
        x = 'Outros Recebíveis Adquiridos'
    
    #Retornar valor de X
    return x

def tempoDivida(y):
    if 'a vencer até 30 dias' in y:
        y = '1. Em até 30 dias'
    elif 'a vencer de 31 a 60 dias' in y:
        y = '2. De 31 a 60 dias'
    elif 'a vencer de 61 a 90 dias' in y:
        y = '3. De 61 a 90 dias'
    elif 'a vencer de 91 a 180 dias' in y:
        y = '4. De 91 a 180 dias'
    else:
        y = '5. Acima de 180'
        
    #Retornar Valor
    return y
```

```python
dict = {'nomeCliente': [], 'categoria':[], 'grupo': [],'tipoDivida': [],'nomeDivida': [], 'valor da dívida': []}

for i in range(len(listaJson[0])):
    nomeDivida = deParaDividas(listaDivida[i])
    tipoDeCredito = listaTipo[i]
    statusAVencer = tempoDivida(listaTipo[i])
    categoriaDivida = listaCategoria[i]
    grupoDivida = listaGrupo[i]
    valorDivida = listaValor[i]
    
    dict['nomeCliente'].append(nome)
    dict['categoria'].append(categoriaDivida)
    dict['tipoDivida'].append(statusAVencer)
    dict['grupo'].append(grupoDivida)
    dict['valor da dívida'].append(valorDivida)
    dict['nomeDivida'].append(nomeDivida)

    #print(f'{nome}: {nomeDivida} - R${valorDivida} | {statusAVencer}')

```

```python
df = pd.DataFrame(dict)
df = df.drop_duplicates()
df.to_excel(r'C:\Users\Matheus\Desktop\df.xlsx')
df
```
